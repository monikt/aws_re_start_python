## Alinear listas de películas
Tenemos tres listas: títulos, años y bandera de "buena película". Vamos a escribir una función para combinarlas y devolver una lista de tuplas o diccionarios, aplicando buenas prácticas.

````python
def alinear_listas(titulos, anios, buenas):
    """Combina tres listas en una lista de diccionarios.

    Parámetros:
        titulos (list of str): títulos de las películas.
        anios (list of int): años de estreno.
        buenas (list of bool): indicador de si es una buena película.

    Retorna:
        list of dict: cada diccionario tiene claves 'titulo', 'anio' y 'esBuena'.

    La función valida que las tres listas tengan la misma longitud y lanza
    ValueError si no es así, de acuerdo con buenas prácticas.
    """
    if not (len(titulos) == len(anios) == len(buenas)):
        raise ValueError("Todas las listas deben tener la misma longitud")

    resultado = []
    for t, a, b in zip(titulos, anios, buenas):
        resultado.append({
            "titulo": t,
            "anio": a,
            "esBuena": b,
        })
    return resultado

# ejemplo de uso
titulos = [
    "The Shawshank Redemption",
    "The Dark Knight",
    "Inception",
    "Pulp Fiction",
    "The Matrix",
    "Interstellar",
    "Parasite",
    "Gladiator",
    "Titanic",
    "The Lord of the Rings: The Fellowship of the Ring",
]
anios = [1994, 2008, 2010, 1994, 1999, 2014, 2019, 2000, 1997, 2001]
buenas = [True, False, True, False, True, False, True, False, True, False]

peliculas = alinear_listas(titulos, anios, buenas)
print(peliculas)
````

## Visualizar datos como tabla
Convertiremos la lista de diccionarios resultante en un DataFrame de pandas y mostraremos una tabla. También añadiremos un gráfico sencillo si se desea.

````python
import pandas as pd
import matplotlib.pyplot as plt

# convertir en DataFrame
df_peliculas = pd.DataFrame(peliculas)

display(df_peliculas)  # muestra la tabla en el notebook

# opcional: graficar año vs titulo como un gráfico de barras
plt.figure(figsize=(12,6))
plt.bar(df_peliculas['titulo'], df_peliculas['anio'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Año de estreno de cada película')
plt.ylabel('Año')
plt.tight_layout()
plt.show()
````
