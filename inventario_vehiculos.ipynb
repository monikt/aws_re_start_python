5. **Preparar la lista de inventario**
    ```python
    myInventoryList = []
    ```
    - Lista vacía donde se irán acumulando copias de `myVehicle` con los datos del CSV.

6. **Leer el archivo CSV y copiar los datos en memoria**
    ```python
    with open('car_fleet.csv') as csvFile:
         csvReader = csv.reader(csvFile, delimiter=',')
         lineCount = 0
         for row in csvReader:
              if lineCount == 0:
                    print(f'Column names are: {", ".join(row)}')
                    lineCount += 1
              else:
                    print(f'vin: {row[0]} make: {row[1]}, model: {row[2]}, '
                            f'year: {row[3]}, range: {row[4]}, topSpeed: {row[5]}, '
                            f'zeroSixty: {row[6]}, mileage: {row[7]}')
                    currentVehicle = copy.deepcopy(myVehicle)
                    currentVehicle["vin"] = row[0]
                    currentVehicle["make"] = row[1]
                    currentVehicle["model"] = row[2]
                    currentVehicle["year"] = row[3]

                    currentVehicle["topSpeed"] = row[5]
                    currentVehicle["zeroSixty"] = row[6]
                    currentVehicle["mileage"] = row[7]
                    myInventoryList.append(currentVehicle)
                    lineCount += 1
         print(f'Processed {lineCount} lines.')
    ```

    - Se emplea `with open` para manejar el archivo.
    - La primera línea del CSV se trata como encabezados.
    - Cada fila se copia en un nuevo diccionario con `copy.deepcopy` para evitar referencias compartidas.
    - Los diccionarios resultantes se añaden a `myInventoryList`.

7. **Imprimir el inventario cargado**
    ```python
    for myCarProperties in myInventoryList:
         for key, value in myCarProperties.items():
    ```

Esta documentación recoge los pasos para crear y ejecutar un programa de Python que lea un inventario de vehículos desde un fichero CSV, almacene los registros en memoria utilizando diccionarios y listas, y muestre los datos.

````python
import csv
import copy
myVehicle = {
    "vin" : "<empty>",
    "make" : "<empty>" ,
    "model" : "<empty>" ,
    "year" : 0,
    "range" : 0,
    "topSpeed" : 0,
    "zeroSixty" : 0.0,
    "mileage" : 0
}
for key, value in myVehicle.items():
    print("{} : {}".format(key,value))
myInventoryList = []
````

````python
with open('car_fleet.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')  
    lineCount = 0  
    for row in csvReader:
        if lineCount == 0:
            print(f'Column names are: {", ".join(row)}')  
            lineCount += 1  
        else:  
            print(f'vin: {row[0]} make: {row[1]}, model: {row[2]}, year: {row[3]}, range: {row[4]}, topSpeed: {row[5]}, zeroSixty: {row[6]}, mileage: {row[7]}')  
            currentVehicle = copy.deepcopy(myVehicle)  
            currentVehicle["vin"] = row[0]  
            currentVehicle["make"] = row[1]  
            currentVehicle["model"] = row[2]  
            currentVehicle["year"] = row[3]  
            currentVehicle["range"] = row[4]  
            currentVehicle["topSpeed"] = row[5]  
            currentVehicle["zeroSixty"] = row[6]  
            currentVehicle["mileage"] = row[7]  
            myInventoryList.append(currentVehicle)  
            lineCount += 1  
    print(f'Processed {lineCount} lines.')
    
    currentVehicle = copy.deepcopy(myVehicle)
    for myCarProperties in myInventoryList:
        for key, value in myCarProperties.items():
            print("{} : {}".format(key,value))
        print("-----")

# convert inventory to pandas dataframe for plotting
import pandas as pd

if myInventoryList:
    df = pd.DataFrame(myInventoryList)
    df['topSpeed'] = pd.to_numeric(df['topSpeed'])
    df['year'] = pd.to_numeric(df['year'])
    print(df[['vin','make','model','year','topSpeed']])

    # plot topSpeed bar chart highlighting max
    import plotly.express as px
    import plotly.io as pio
    # choose a renderer that doesn't require nbformat, fallback to HTML
    pio.renderers.default = 'notebook_connected'
    max_speed = df['topSpeed'].max()
    df['highlight'] = df['topSpeed'] == max_speed
    fig = px.bar(df, x='vin', y='topSpeed', color='highlight',
                 color_discrete_map={True: 'red', False: 'steelblue'},
                 title='Top Speed by Vehicle (max highlighted in red)')
    try:
        fig.show()
    except Exception as e:
        print("plotly show failed, rendering inline via HTML\n", e)
        from IPython.display import HTML
        display(HTML(fig.to_html(include_plotlyjs='cdn')))
````
